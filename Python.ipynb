#Solution for question 1

```python
def find_highest_frequency_word_length(string):
    word_list = string.split()
    word_count = {}

    for word in word_list:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1

    highest_frequency = max(word_count.values())
    highest_frequency_words = [word for word, count in word_count.items() if count == highest_frequency]
    highest_frequency_word_length = len(max(highest_frequency_words, key=len))

    return highest_frequency_word_length
```

```python
string = "this is a test string with some repeated words this is a test"
print(find_highest_frequency_word_length(string))
```

#Solution for question 2

```python
from collections import Counter
def isValidString(s):
    char_count = Counter(s)  # Count the frequency of each character

    # Count the frequencies of the character frequencies
    freq_count = Counter(char_count.values())

    if len(freq_count) == 1:
        return "YES"

    if len(freq_count) > 2:
        return "NO"

    (freq1, count1), (freq2, count2) = freq_count.items()

    if (freq1 == 1 and count1 == 1) or (freq2 == 1 and count2 == 1):
        return "YES"

    if (freq1 - freq2 == 1 and count1 == 1) or (freq2 - freq1 == 1 and count2 == 1):
        return "YES"

    return "NO"
```

```python
s = "aabbccdd"
print(isValidString(s))
```

#Solution for question 4

```python
import csv
import requests

def download_data(url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.text
    else:
        raise Exception("Failed to download data from the provided URL.")

def process_data(data):
    # Perform data processing operations here
    # In this example, we split the data by newline and then by comma
    processed_data = [line.split(',') for line in data.split('\n')]
    return processed_data

def save_as_csv(data, filename):
    with open(filename, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerows(data)
    print(f"Data successfully saved as {filename}")
```

```python
# Example usage
url = "https://data.nasa.gov/resource/y77d-th95.json"
data = download_data(url)
processed_data = process_data(data)
save_as_csv(processed_data, "output.csv")
```

```python

import requests
import re

# API endpoint
api_url = "http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes"

# Send GET request to the API endpoint
response = requests.get(api_url)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()
    
    # Extract show information
    show_name = data["name"]
    print("Show Name:", show_name)
    
    # Extract episode information
    episodes = data["_embedded"]["episodes"]
    print("\nEpisode List:")
    for episode in episodes:
        episode_id = episode["id"]
        episode_url = episode["url"]
        episode_name = episode["name"]
        episode_season = episode["season"]
        episode_number = episode["number"]
        episode_type = episode["type"]
        episode_airdate = episode["airdate"]
        episode_airtime = episode["airtime"]
        episode_runtime = episode["runtime"]
        episode_rating = episode["rating"]["average"]
        episode_summary = re.sub('<.*?>', '', episode["summary"])
        episode_image_medium = episode["image"]["medium"]
        episode_image_original = episode["image"]["original"]
        
        print("\nEpisode ID:", episode_id)
        print("Episode URL:", episode_url)
        print("Episode Name:", episode_name)
        print("Season:", episode_season)
        print("Episode Number:", episode_number)
        print("Type:", episode_type)
        print("Airdate:", episode_airdate)
        print("Airtime:", episode_airtime)
        print("Runtime:", episode_runtime)
        print("Average Rating:", episode_rating)
        print("Summary:", episode_summary)
        print("Medium Image Link:", episode_image_medium)
        print("Original Image Link:", episode_image_original)
else:
    print("Error:", response.status_code)
```

#Solution for question 6

```python

import pandas as pd
import requests
import matplotlib.pyplot as plt

# Fetch the data from the URL
url = "https://raw.githubusercontent.com/Biuni/PokemonGO-Pokedex/master/pokedex.json"
response = requests.get(url)
data = response.json()

# Convert the data into a pandas DataFrame
df = pd.DataFrame(data["pokemon"])

# 1. Get all Pokemons whose spawn rate is less than 5%
spawn_rate_threshold = 5
df_less_than_5_percent_spawn_rate = df[df["spawn_chance"] < spawn_rate_threshold]
print("Pokemons with spawn rate less than 5%:")
print(df_less_than_5_percent_spawn_rate[["name", "spawn_chance"]])

# Plot the spawn rate distribution
plt.hist(df["spawn_chance"], bins=20)
plt.xlabel("Spawn Rate (%)")
plt.ylabel("Count")
plt.title("Distribution of Pokemon Spawn Rates")
plt.show()

# 2. Get all Pokemons that have less than 4 weaknesses
max_weaknesses = 4
df_less_than_4_weaknesses = df[df["weaknesses"].apply(lambda x: len(x)) < max_weaknesses]
print("Pokemons with less than 4 weaknesses:")
print(df_less_than_4_weaknesses[["name", "weaknesses"]])

# 3. Get all Pokemons that have no multipliers at all
df_no_multipliers = df[df["multipliers"].apply(lambda x: len(x) if x is not None else 0) == 0]
print("Pokemons with no multipliers:")
print(df_no_multipliers[["name", "multipliers"]])

# 4. Get all Pokemons that do not have more than 2 evolutions
max_evolutions = 2
df_less_than_2_evolutions = df[df["next_evolution"].apply(lambda x: len(x) if isinstance(x, list) else 0) <= max_evolutions]
print("Pokemons with 2 or fewer evolutions:")
print(df_less_than_2_evolutions[["name", "next_evolution"]])

# 5. Get all Pokemons whose spawn time is less than 300 seconds
max_spawn_time = 300
df_less_than_300_spawn_time = df[df["spawn_time"].apply(lambda x: int(x.split(":")[0]) * 60 + int(x.split(":")[1]) if x != "N/A" else float("inf")) < max_spawn_time]
print("Pokemons with spawn time less than 300 seconds:")
print(df_less_than_300_spawn_time[["name", "spawn_time"]])

# 6. Get all Pokemon who have more than two types of capabilities
min_capabilities = 2
df_more_than_2_capabilities = df[df["type"].apply(lambda x: len(x)) > min_capabilities]
print("Pokemons with more than two types of capabilities:")
print(df_more_than_2_capabilities[["name", "type"]])
```

```python

import pandas as pd
import matplotlib.pyplot as plt

# Read the JSON data into a pandas DataFrame
data_url = 'https://data.nasa.gov/resource/y77d-th95.json'
df = pd.read_json(data_url)

# Convert the 'year' column to datetime format
df['year'] = pd.to_datetime(df['year'], errors='coerce')

# Check the column names in the DataFrame
print(df.columns)

# Output:
# Index(['name', 'id', 'nametype', 'recclass', 'mass (g)', 'fall', 'year', 'reclat', 'reclong', 'geolocation'], dtype='object')

# Update the column name if necessary
df.rename(columns={'mass (g)': 'mass'}, inplace=True)
# Question 1: Get all the Earth meteorites that fell before the year 2000
earth_meteorites_before_2000 = df[(df['reclat'].notnull()) & (df['year'].dt.year < 2000)]
print(earth_meteorites_before_2000[['name', 'year']])

# Question 2: Get all the Earth meteorites' coordinates that fell before the year 1970
earth_meteorites_coords_before_1970 = df[(df['reclat'].notnull()) & (df['year'].dt.year < 1970)][['reclat', 'reclong']]
print(earth_meteorites_coords_before_1970)

# Question 3: Get all Earth meteorites whose mass was more than 10000kg
earth_meteorites_mass_above_10000kg = df[(df['reclat'].notnull()) & (df['mass'] > 10000)]
print(earth_meteorites_mass_above_10000kg[['name', 'mass']])
```

#Solution for question 8

```python

import requests
import pandas as pd
import matplotlib.pyplot as plt

# Fetch data from the API
response = requests.get('http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes')
data = response.json()

# Create a DataFrame to store the episode data
episodes_df = pd.DataFrame(data['_embedded']['episodes'])

# Convert 'rating' column to numeric type
episodes_df['rating'] = pd.to_numeric(episodes_df['rating'], errors='coerce')

# Get all the overall ratings for each season
season_ratings = episodes_df.groupby('season')['rating'].mean()

# Plotting the ratings for all the seasons
plt.bar(season_ratings.index, season_ratings.values)
plt.xlabel('Season')
plt.ylabel('Average Rating')
plt.title('Average Ratings for Each Season')
plt.show()

# Get all the episode names whose average rating is more than 8 for every season
high_rated_episodes = episodes_df.groupby(['season', 'name']).filter(lambda x: x['rating'].mean() > 8)['name'].unique()

# Get all the episode names that aired before May 2019
episodes_before_may_2019 = episodes_df[episodes_df['airdate'] < '2019-05']

# Get the episode name from each season with the highest and lowest rating
highest_rated_episodes = episodes_df.groupby('season')['rating'].idxmax().dropna().astype(int)
lowest_rated_episodes = episodes_df.groupby('season')['rating'].idxmin().dropna().astype(int)

if highest_rated_episodes.empty:
    highest_rated_names = pd.DataFrame(columns=['season', 'name'])
else:
    highest_rated_names = episodes_df.loc[highest_rated_episodes, ['season', 'name']]

if lowest_rated_episodes.empty:
    lowest_rated_names = pd.DataFrame(columns=['season', 'name'])
else:
    lowest_rated_names = episodes_df.loc[lowest_rated_episodes, ['season', 'name']]

# Get the summary for the most popular (ratings) episode in every season
most_popular_episodes = episodes_df.groupby('season')['rating'].idxmax().dropna().astype(int)
most_popular_summaries = episodes_df.loc[most_popular_episodes, ['season', 'name', 'summary']]

if most_popular_episodes.empty:
    most_popular_summaries = pd.DataFrame(columns=['season', 'name', 'summary'])

# Print the results
print("Episode names with average rating > 8 for every season:")
print(high_rated_episodes)

print("\nEpisode names that aired before May 2019:")
print(episodes_before_may_2019['name'])

print("\nEpisode name from each season with the highest rating:")
print(highest_rated_names)

print("\nEpisode name from each season with the lowest rating:")
print(lowest_rated_names)

print("\nSummary for the most popular episode in every season:")
print(most_popular_summaries)
```

#Solution for question 9

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://data.wa.gov/api/views/f6w7-q2d2/rows.csv?accessType=DOWNLOAD"
df = pd.read_csv(url)

# Check the available column names
print(df.columns)

# Get all the cars and their types that do not qualify for clean alternative fuel vehicle
non_clean_cars = df[df["Clean Alternative Fuel Vehicle (CAFV) Eligibility"] == "No"]
car_types = non_clean_cars[["Make", "Model"]]
print("Cars that do not qualify for clean alternative fuel vehicle:")
print(car_types)


# Get all TESLA cars with the model year and model type made in Bothell City
tesla_cars = df[(df["Make"] == "TESLA") & (df["City"] == "Bothell")]
tesla_details = tesla_cars[["Model Year", "Model"]]
print("TESLA cars with the model year and model type made in Bothell City:")
print(tesla_details)

# Get all the cars that have an electric range of more than 100 and were made after 2015
electric_cars = df[(df["Electric Range"] > 100) & (df["Model Year"] > 2015)]
print("Cars with an electric range of more than 100 and made after 2015:")
print(electric_cars)

# Draw plots to show the distribution between city and electric vehicle type
df.groupby(["City", "Electric Vehicle Type"]).size().unstack().plot(kind="bar", stacked=True)
plt.xlabel("City")
plt.ylabel("Count")
plt.title("Distribution of City and Electric Vehicle Type")
plt.show()
```

#Solution for question 10

```python


import spacy
from collections import Counter

def count_pos_tags(text):
    # Load the spaCy English language model
    #en_core_web_sm is a pre-trained model for English language processing
    nlp = spacy.load('en_core_web_sm')
    
    # Process the text
    doc = nlp(text)
    
    # Initialize counts for each part of speech
    #Counter object where the keys are the part-of-speech tags (e.g., 'NOUN', 'VERB', 'PRON', 'ADJ') 
    #retrieves the part-of-speech tag of each token in the document. pos
    counts = Counter([token.pos_ for token in doc])
    
    # Create a dictionary with the part of speech counts
    pos_counts = {'Noun': counts['NOUN'], 'Verb': counts['VERB'], 'Pronoun': counts['PRON'], 'Adjective': counts['ADJ']}
    
    return pos_counts
```

```python
text ="Books are our best friends and the best way to know about the world. Reading books helps us change our perspective of the world. We have listened to a lot of stories from our elders, but with time, most of our families became limited to only a father, a mother and a child. Gone are those days when grandparents used to sleep beside their grandchildren and narrate bedtime stories"

counts = count_pos_tags(text)
print(counts)
```
